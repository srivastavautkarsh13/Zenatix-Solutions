Its clear from the data that Power Consumption is available for the months of August and September.
Also, data is recorded after every two minutes gap, thats why half of the values are NULL.

```python
df['datetime'] = pd.to_datetime(df['0'])
```

```python
df.columns
```

```python
df['month'] = df['datetime'].apply(lambda x : x.month)
df['day'] = df['datetime'].apply(lambda x : x.day)
df['hour'] = df['datetime'].apply(lambda x : x.hour)
df['minute'] = df['datetime'].apply(lambda x : x.minute)
```

```python
df.head()
```

```python
df.drop(['0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns
```

# Now, We have recordings for every minute and half of the data is NULL.

# Since we have half of the dataset values as NULL, we can eitrher drop it  or impute the values. Now imputing means we would have to replace half of dataset values which will be like adding a lot of information. So lets drop all the NULL values.

```python
#keeping it as a backup of original data
new_df = df.copy()
```

```python
new_df.shape
```

```python
new_df = new_df.dropna(axis=0)
```

```python
ac_columns = ['AC 1', 'AC 2', 'AC 3', 'AC 4', 'AC 5', 'AC 6', 'AC 7', 'AC 8', 'AC 9',
       'AC 10', 'AC 11', 'AC 12', 'AC 13', 'AC 14', 'AC 15', 'AC 16', 'AC 17',
       'AC 18']
```

```python
day_hour_df = df.groupby(['month','day','hour'],as_index=False)['AC 1', 'AC 2', 'AC 3', 'AC 4', 'AC 5', 'AC 6', 'AC 7', 'AC 8', 'AC 9',
       'AC 10', 'AC 11', 'AC 12', 'AC 13', 'AC 14', 'AC 15', 'AC 16', 'AC 17',
       'AC 18'].sum()
```

```python
day_hour_df.head()
```

```python
#sns.boxplot(data=hour_df)
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x=ac_columns,y=day_hour_df[ac_columns].describe().loc['mean'],orient='v')
plt.ylabel('watts')
plt.title('Average Power consumption of ACs for August and September months')
```

```python
plt.figure(figsize=(20,10))
count = 1
for ac in ac_columns:
    plt.subplot(6,3,count)
    count += 1
    sns.lineplot(x='hour',y=ac,data=day_hour_df,hue='month')
```
