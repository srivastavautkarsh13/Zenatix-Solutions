**Now Imputing all the NULL values with average of previos and next values**

```python
new_df = new_df.ffill().bfill()
```

```python
new_df.shape
```

```python
(new_df.isnull().sum()/len(df))*100
```

```python
new_df['datetime'] = pd.to_datetime(new_df['0'])
```

```python
new_df.columns
```

```python
new_df['month'] = new_df['datetime'].apply(lambda x : x.month)
new_df['day'] = new_df['datetime'].apply(lambda x : x.day)
new_df['hour'] = new_df['datetime'].apply(lambda x : x.hour)
new_df['minute'] = new_df['datetime'].apply(lambda x : x.minute)
```

```python
new_df.drop(['0'], axis=1, inplace=True)
```

```python
new_df.head()
```

```python
rep_list = set(new_df['datetime'].value_counts() > 1)
rep_list
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=new_df.iloc[:,:-4])
```

```python
new_df.describe()
```

```python
ac_columns = ['AC 1', 'AC 2', 'AC 3', 'AC 4', 'AC 5', 'AC 6', 'AC 7', 'AC 8', 'AC 9',
       'AC 10', 'AC 11', 'AC 12', 'AC 13', 'AC 14', 'AC 15', 'AC 16', 'AC 17',
       'AC 18']
```

```python
new_df.tail()
```

```python
hour_df_sum = new_df.groupby(['month','day','hour'],as_index=False)[ac_columns].sum()
hour_df_mean = new_df.groupby(['month','day','hour'],as_index=False)[ac_columns].mean()
hour_df_std = new_df.groupby(['month','day','hour'],as_index=False)[ac_columns].std()
```

```python
hour_df_sum.tail()
```

```python
hour_df_mean.head()
```

```python
hour_df_std.tail()
```

```python
plt.figure(figsize=(30,20))
count = 1
for ac in ac_columns:
    plt.subplot(6,3,count)
    count += 1
    sns.barplot(x='hour',y=ac,data=hour_df_sum,hue='month')
```

```python
plt.figure(figsize=(30,20))
count = 1
for ac in ac_columns:
    plt.subplot(6,3,count)
    count += 1
    sns.lineplot(x='hour',y=ac,data=hour_df_mean,hue='month',palette="flare")
```

```python
plt.figure(figsize=(30,20))
count = 1
for ac in ac_columns:
    plt.subplot(6,3,count)
    count += 1
    sns.lineplot(x='hour',y=ac,data=hour_df_std,hue='month',palette="flare")
```
